## The dataset contains 15,000 entries and 14 columns. Here's a summary of the columns:

Product_Name: Name of the cosmetic product (string).

Brand: Brand associated with the product (string).

Category: Type or category of the product (e.g., Blush, Highlighter) (string).

Usage_Frequency: Frequency of use (e.g., Daily, Weekly, Occasional) (string).

Price_USD: Price of the product in USD (float).

Rating: Average customer rating of the product (float).

Number_of_Reviews: Number of customer reviews (integer).

Product_Size: Size of the product (e.g., 30ml, 100ml) (string).

Skin_Type: Type of skin the product is suitable for (e.g., Sensitive, Oily) (string).

Gender_Target: Gender targeted by the product (e.g., Female, Male, Unisex) (string).

Packaging_Type: Type of packaging (e.g., Tube, Bottle, Compact) (string).

Main_Ingredient: Primary ingredient in the product (e.g., Retinol, Aloe Vera) (string).

Cruelty_Free: Indicates whether the product is cruelty-free (boolean).

Country_of_Origin: Country where the product is manufactured (string).

```python
# Create a function to calculate the average price range for beauty brands
def calculate_average_price_range(df):
    # Check if 'Brand' column exists
    if 'Brand' not in df.columns:
        raise ValueError("DataFrame must contain a 'Brand' column.")
        
    # Group by brand and calculate average price
    average_price_data = df.groupby('Brand')['Price_USD'].agg(['min', 'max']).reset_index()
    average_price_data['average_price'] = (average_price_data['min'] + average_price_data['max']) / 2
    return average_price_data[['Brand', 'average_price']]  # Corrected to return both columns

# Function to sort the brands by their average price range
def sort_brands_by_average_price(average_price_data):
    return average_price_data.sort_values(by='average_price')

# Calculate the average price range for each brand
average_price_data = calculate_average_price_range(df)

# Sort the brands by their average price range
sorted_brands = sort_brands_by_average_price(average_price_data)

# Print the sorted list of beauty brands with their average price ranges
for index, brand in sorted_brands.iterrows():
    print(f"{brand['Brand']}: ${brand['average_price']:.2f}")  # Corrected the indexing to 'Brand'
```
